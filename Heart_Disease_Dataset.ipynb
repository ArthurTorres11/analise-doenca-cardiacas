```python
# Renomeando as colunas
dados.rename(columns={
    'age': 'idade',
    'sex': 'sexo',
    'cp': 'tipo_de_dor_peito',
    'trestbps': 'pressao_arterial',
    'chol': 'colesterol',
    'fbs': 'glicemia_jejum',
    'restecg': 'resul_eletro',
    'thalach': 'freq_max_repouso',
    'exang': 'angina',
    'oldpeak': 'depressao_st',
    'slope': 'inclinacao_st',
    'ca': 'vasos_fluorosopia',
    'thal': 'thal'
}, inplace=True)
```

```python
dados.head()
```

### Descrição das Colunas:

- **`idade`**: Idade do paciente.
- **`sexo`**: Sexo do paciente (0 = masculino, 1 = feminino).
- **`tipo_de_dor_peito`**: Tipo de dor no peito
  - 0 = angina típica
  - 1 = angina atípica
  - 2 = dor não-anginosa
  - 3 = sem dor
- **`pressao_arterial`**: Pressão arterial em repouso (mmHg).
- **`colesterol`**: Colesterol sérico (mg/dl).
- **`glicemia_jejum`**: Glicemia em jejum
  - 0 = <= 120 mg/dl
  - 1 = > 120 mg/dl
- **`resul_eletro`**: Resultado do eletrocardiograma
  - 0 = normal
  - 1 = anormalidade de onda ST-T
  - 2 = hipertrofia ventricular esquerda
- **`freq_max_repouso`**: Frequência cardíaca máxima atingida durante o exercício.
- **`angina`**: Angina induzida por exercício
  - 0 = não
  - 1 = sim
- **`depressao_st`**: Depressão do segmento ST no exercício.
- **`inclinacao_st`**: Inclinação do pico do segmento ST.
- **`vasos_fluorosopia`**: Número de vasos coloridos pela fluoroscopia (0-3).
- **`thal`**: Condição da thalassemia
  - 0 = normal
  - 1 = defeito corrigido
  - 2 = defeito reversível

```python
## Separando se o sexo pode ser um fator de risco:
dados_homens = dados.query('sexo == 0')
dados_mulheres = dados.query('sexo == 1')
```

```python
contagem_sexo = dados['sexo'].value_counts()
```

```python
plt.bar(contagem_sexo.index, contagem_sexo.values, color=['blue', 'red'])
plt.title('Distribuição de Sexo', fontsize=14)
plt.xlabel('Sexo', fontsize=12)
plt.ylabel('Número de Pacientes', fontsize=12)
plt.xticks([0, 1], ['Homens', 'Mulheres'])
plt.show
```

```python
## Separando agravantes entre os sexos
media_homens = dados_homens[['idade', 'pressao_arterial', 'colesterol']].mean()
```

```python
media_mulheres = dados_mulheres[['idade', 'pressao_arterial', 'colesterol']].mean()
```

```python
media = pd.DataFrame({
    'homens': media_homens,
    'mulheres': media_mulheres
})

media.plot(kind='bar', figsize=(8, 6), color=['blue', 'red'])

plt.title('Comparação de Médias entre Homens e Mulheres', fontsize=14)
plt.ylabel('Valor Médio', fontsize=12)
plt.xlabel('Variáveis', fontsize=12)

plt.xticks(rotation=0)
plt.show()
```
